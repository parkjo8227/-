```python
import csv
import matplotlib.pyplot as plt

#seoul.csv 불러오기
f= open('seoul.csv')
data=csv.reader(f)
next(data)

#일별 데이터 저장위한 list생성
jan=[]
feb=[]
mar=[]
apr=[]
may=[]
jun=[]
jul=[]
aug=[]
sep=[]
ten=[]
nov=[]
dec=[]

#31일있는 월의 일별데이터저장
for i in range(31):
    jan.append([])
    mar.append([])
    may.append([])
    jul.append([])
    aug.append([])
    ten.append([])
    dec.append([])

#30일있는 월의 일별데이터저장
for i in range(30):
    apr.append([])
    jun.append([])
    sep.append([])
    nov.append([])

#28~29일있는 월의 일별데이터저장
for i in range(29):
    feb.append([])

    
for row in data:
    if row[-1] !='' and row[-2]:
        if int(row[0].split('-')[0])>=1999:              # 1999년이후 데이터
            if row[-1] !='':
                m=row[0].split('-')[1]
                if m == '01':                           # 1월 데이터
                    d=int(row[0].split('-')[2])         # 날짜항목에서 일만 추출
                    jan[d-1].append(float(row[-1]))
                if m == '02':                           # 2월 데이터
                    d=int(row[0].split('-')[2])
                    feb[d-1].append(float(row[-1]))
                if m == '03':                           # 3월 데이터
                    d=int(row[0].split('-')[2])
                    mar[d-1].append(float(row[-1]))
                if m == '04':                           # 4월 데이터
                    d=int(row[0].split('-')[2])
                    apr[d-1].append(float(row[-1]))
                if m == '05':                           # 5월 데이터
                    d=int(row[0].split('-')[2])
                    may[d-1].append(float(row[-1]))
                if m == '06':                           # 6월 데이터
                    d=int(row[0].split('-')[2])
                    jun[d-1].append(float(row[-1]))
                if m == '07':                           # 7월 데이터
                    d=int(row[0].split('-')[2])
                    jul[d-1].append(float(row[-1]))
                if m == '08':                           # 8월 데이터
                    d=int(row[0].split('-')[2])
                    aug[d-1].append(float(row[-1]))
                if m == '09':                           # 9월 데이터
                    d=int(row[0].split('-')[2])
                    sep[d-1].append(float(row[-1]))
                if m == '10':                           # 10월 데이터
                    d=int(row[0].split('-')[2])
                    ten[d-1].append(float(row[-1]))
                if m == '11':                           # 11월 데이터
                    d=int(row[0].split('-')[2])
                    nov[d-1].append(float(row[-1]))
                if m == '12':                           # 12월 데이터
                    d=int(row[0].split('-')[2])
                    dec[d-1].append(float(row[-1]))
 
#

#그래프 생성
plt.rc('axes', unicode_minus=False)

plt.figure(figsize=(10,5),dpi=100)
plt.title('1999년부터 20년간 1월 최고 기온 일별 변화')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(jan,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999년부터 20년간 2월 최고 기온 일별 변화')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(feb,showfliers=False)


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999년부터 20년간 3월 최고 기온 일별 변화')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(mar,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999년부터 20년간 4월 최고 기온 일별 변화')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(apr,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999년부터 20년간 5월 최고 기온 일별 변화')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(may,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999년부터 20년간 6월 최고 기온 일별 변화')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(jun,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999년부터 20년간 7월 최고 기온 일별 변화')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(jul,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999년부터 20년간 8월 최고 기온 일별 변화')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(aug,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999년부터 20년간 9월 최고 기온 일별 변화')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(sep,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999년부터 20년간 10월 최고 기온 일별 변화')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(ten,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999년부터 20년간 11월 최고 기온 일별 변화')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(nov,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999년부터 20년간 12월 최고 기온 일별 변화')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(dec,showfliers=False)
plt.show()
```
