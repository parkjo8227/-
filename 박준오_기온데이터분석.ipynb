```python
import csv
import matplotlib.pyplot as plt

#seoul.csv 
f= open('seoul.csv')
data=csv.reader(f)
next(data)

#   list
jan=[]
feb=[]
mar=[]
apr=[]
may=[]
jun=[]
jul=[]
aug=[]
sep=[]
ten=[]
nov=[]
dec=[]

#31  
for i in range(31):
    jan.append([])
    mar.append([])
    may.append([])
    jul.append([])
    aug.append([])
    ten.append([])
    dec.append([])

#30  
for i in range(30):
    apr.append([])
    jun.append([])
    sep.append([])
    nov.append([])

#28~29  
for i in range(29):
    feb.append([])

    
for row in data:
    if row[-1] !='' and row[-2]:
        if int(row[0].split('-')[0])>=1999:              # 1999 
            if row[-1] !='':
                m=row[0].split('-')[1]
                if m == '01':                           # 1 
                    d=int(row[0].split('-')[2])         #   
                    jan[d-1].append(float(row[-1]))
                if m == '02':                           # 2 
                    d=int(row[0].split('-')[2])
                    feb[d-1].append(float(row[-1]))
                if m == '03':                           # 3 
                    d=int(row[0].split('-')[2])
                    mar[d-1].append(float(row[-1]))
                if m == '04':                           # 4 
                    d=int(row[0].split('-')[2])
                    apr[d-1].append(float(row[-1]))
                if m == '05':                           # 5 
                    d=int(row[0].split('-')[2])
                    may[d-1].append(float(row[-1]))
                if m == '06':                           # 6 
                    d=int(row[0].split('-')[2])
                    jun[d-1].append(float(row[-1]))
                if m == '07':                           # 7 
                    d=int(row[0].split('-')[2])
                    jul[d-1].append(float(row[-1]))
                if m == '08':                           # 8 
                    d=int(row[0].split('-')[2])
                    aug[d-1].append(float(row[-1]))
                if m == '09':                           # 9 
                    d=int(row[0].split('-')[2])
                    sep[d-1].append(float(row[-1]))
                if m == '10':                           # 10 
                    d=int(row[0].split('-')[2])
                    ten[d-1].append(float(row[-1]))
                if m == '11':                           # 11 
                    d=int(row[0].split('-')[2])
                    nov[d-1].append(float(row[-1]))
                if m == '12':                           # 12 
                    d=int(row[0].split('-')[2])
                    dec[d-1].append(float(row[-1]))
 
#

# 
plt.rc('axes', unicode_minus=False)

plt.figure(figsize=(10,5),dpi=100)
plt.title('1999 20 1    ')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(jan,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999 20 2    ')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(feb,showfliers=False)


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999 20 3    ')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(mar,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999 20 4    ')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(apr,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999 20 5    ')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(may,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999 20 6    ')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(jun,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999 20 7    ')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(jul,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999 20 8    ')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(aug,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999 20 9    ')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(sep,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999 20 10    ')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(ten,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999 20 11    ')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(nov,showfliers=False)
plt.show()


plt.figure(figsize=(10,5),dpi=100)
plt.title('1999 20 12    ')
plt.style.use('ggplot')
plt.rc('font',family='malgun Gothic')
plt.boxplot(dec,showfliers=False)
plt.show()
```
