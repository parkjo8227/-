## 셜록홈즈_바스커빌가의개


```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
with open("The_Hound_of_the_Baskervilles.txt") as f:
    raw_text=f.readlines()

data = []
STOPWORDS = ['\n','있는','한','있었다.','있었다.\n','그','이','그리고','그는','그의','수','그렇게','그러자','두','말했다.\n','본','좀','그들은','말했다.','것','그가','걸']
fillterd_data = []
f_data = []

for x in raw_text:
    line = x.split(" ")
    for w in line:
        data.append(w)

for tag in data:
    if len(tag) >=1:
        if tag not in STOPWORDS:
            fillterd_data.append(tag)
            
from collections import Counter
cnt_fillterd_data = Counter(fillterd_data)
cnt_fillterd_data.most_common(200)
f_data = cnt_fillterd_data.most_common(50)
 
df = pd.DataFrame(f_data)

plt.rc('font', family='malgun Gothic')
plt.rcParams['axes.unicode_minus']=False

df.columns = ['Text','cnt']
df.set_index('Text',inplace=True)

te = df.plot(kind='bar',title="단어 별 빈도수",figsize = (12,4), legend = True)
te.set_xlabel('단어')
te.set_ylabel('빈도수')
te.legend(['단어'])
plt.show()

korean_font_path = 'C:\Windows/Fonts/malgunbd.ttf'
wordcloud_image = WordCloud(background_color='white',font_path = korean_font_path, max_words=200, relative_scaling=0.3,max_font_size=600,width=2000,height=1000). generate_from_frequencies(cnt_fillterd_data)
image_file_name=('The_Hound_of_the_Baskervilles.png')
wordcloud_image.to_file(image_file_name)
plt.imshow(wordcloud_image)
plt.axis("off")
plt.show()

```
