# 노령화 정도


```python
import csv
import matplotlib.pyplot as plt


f = open('age.csv')
data=csv.reader(f)

eve=[]
yo=0
ol=0

name=input('고령화 비율을 알고싶은 지역의 이름을 입력하세요: ')

for row in data:
    if name in row[0]:
        for i in row[3:]:
            eve.append(int(i.replace(',','')))
        for i in row[3:68]:
            yo=yo+int(i.replace(',',''))
        for i in row[69:]:
            ol=ol+int(i.replace(',',''))
            
        axx = yo + ol
            
plt.style.use('ggplot')
plt.rc('font', family='malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title(name +' 의 인구 구조')
plt.plot(eve)
plt.show()

plt.title(name +' 의 고령화 비율')
plt.pie([yo,ol],labels=['65세 미만','65세 이상'],autopct='%.1f%%',explode=(0,0.1))
plt.legend()
plt.show()

if (ol/axx*100) >=20:
    print(name + "은 20% 이상으로 초고령 사회입니다")
elif (ol/axx*100) >=14:
    print(name + "은 14% 이상으로 고령 사회입니다")
elif (ol/axx*100) >=7:
    print(name + "은 7% 이상으로 고령화 사회입니다")
```

# 학령인구


```python
import csv
import matplotlib.pyplot as plt


f = open('age.csv')
data=csv.reader(f)

student=0
thing=0
baby=0
sum=[]
print("학령인구는 6세부터 21세입니다")
name=input('학령인구 비율을 알고싶은 지역의 이름을 입력하세요: ')

for row in data:
    if name in row[0]:
        for i in row[3:8]:                            #0세~5세까지
            baby=baby+int(i.replace(',',''))
        for i in row[9:24]:                           #6세~21세까지
            student=student+int(i.replace(',',''))
        for i in row[25:]:                            #22세~
            thing=thing+int(i.replace(',',''))
        sum = thing +baby                            #학령인구를 제외한 나머지 인구

plt.style.use('ggplot')
plt.rc('font', family='malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title(name +' 의 학령인구 비율')
plt.pie([student,sum],labels=['학령인구','나머지인구'],autopct='%.1f%%',explode=(0.1,0))
plt.legend()
plt.show()
```

# 취업가능 인구 정도


```python
import csv
import matplotlib.pyplot as plt


f = open('age.csv')
data=csv.reader(f)

nodong=0
rest=0
no=0
sum1=[]

name=input('취업가능 인구정도 비율을 알고싶은 지역의 이름을 입력하세요: ')

for row in data:
    if name in row[0]:
        for i in row[3:17]:                            #0세~14세까지
            no=no+int(i.replace(',',''))
        for i in row[18:67]:                           #15세~64세까지
            nodong=nodong+int(i.replace(',',''))
        for i in row[68:]:                             #65세~
            rest=rest+int(i.replace(',',''))
        sum1 = no + rest                               #취업가능 인구를 제외한 나머지 인구

print("\n")
print("생산가능인구는 만 15세 ~ 만 64세")
plt.style.use('ggplot')
plt.rc('font', family='malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title(name +' 의 취업가능 인구 비율')
plt.pie([nodong,sum1],labels=['취업가능인구','나머지인구'],autopct='%.1f%%',explode=(0.1,0))
plt.legend()
plt.show()
```

# 연령대별 남녀 성별 비율

```python
import csv
import matplotlib.pyplot as plt


f = open('gender.csv')
data=csv.reader(f)

result=[]
m=[]
f=[]

name=input('연령대별 남녀 성별 비율을 알고 싶은 지역을 입력하세요: ')

for row in data:
    if name in row[0]:
        for i in row[3:104]:                            #남
            m.append(int(i.replace(',','')))
        for i in row[106:]:                            #여
            f.append(int(i.replace(',','')))
        break

plt.rc('font', family='malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title(name +' 의 연령대별 남녀 성별 비율')
plt.plot(m,'red',label='male')
plt.plot(f,'blue',label='female')
plt.legend()
plt.show()
```
